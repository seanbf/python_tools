```python
# Create dataframe from all_files csv's
input_file = list(Button_Upload.value.values())[0]
content = input_file['content']
content = io.StringIO(content.decode('utf-8'))
df = pd.read_csv(content)

print('User: '+str(User))
print('Date Generated: '+str(Date))
print('Files: ')
        
def UpdateSignals(DropDown_Signal1, DropDown_Signal2,DropDown_Signal3,	DropDown_Signal4,	DropDown_Signal5,	DropDown_Signal6,	DropDown_Signal7,	DropDown_Signal8,	DropDown_Signal9,	TextBox_Label1,	TextBox_Label2,	TextBox_Label3,	TextBox_Label4,	TextBox_Label5,	TextBox_Label6,	TextBox_Label7,	TextBox_Label8,	TextBox_Label9,	Toggle_Axis1,	Toggle_Axis2,	Toggle_Axis3,	Toggle_Axis4,	Toggle_Axis5,	Toggle_Axis6,	Toggle_Axis7,	Toggle_Axis8,	Toggle_Axis9,	ColorPicker_Color1,	ColorPicker_Color2,	ColorPicker_Color3,	ColorPicker_Color4,	ColorPicker_Color5,	ColorPicker_Color6,	ColorPicker_Color7,	ColorPicker_Color8,	ColorPicker_Color9,	Toggle_Type1,	Toggle_Type2,	Toggle_Type3,	Toggle_Type4,	Toggle_Type5,	Toggle_Type6,	Toggle_Type7,	Toggle_Type8,	Toggle_Type9):
    Signal_1 = dict([('Signal', DropDown_Signal1.value), ('Name', TextBox_Label1.value), ('Axis', Toggle_Axis1.value), ('Color', ColorPicker_Color1.value), ('Type', Toggle_Type1.value)])
    Signal_2 = dict([('Signal', DropDown_Signal2.value), ('Name', TextBox_Label2.value), ('Axis', Toggle_Axis2.value), ('Color', ColorPicker_Color2.value), ('Type', Toggle_Type2.value)])
    Signal_3 = dict([('Signal', DropDown_Signal3.value), ('Name', TextBox_Label3.value), ('Axis', Toggle_Axis3.value), ('Color', ColorPicker_Color3.value), ('Type', Toggle_Type3.value)])
    Signal_4 = dict([('Signal', DropDown_Signal4.value), ('Name', TextBox_Label4.value), ('Axis', Toggle_Axis4.value), ('Color', ColorPicker_Color4.value), ('Type', Toggle_Type4.value)])
    Signal_5 = dict([('Signal', DropDown_Signal5.value), ('Name', TextBox_Label5.value), ('Axis', Toggle_Axis5.value), ('Color', ColorPicker_Color5.value), ('Type', Toggle_Type5.value)])
    Signal_6 = dict([('Signal', DropDown_Signal6.value), ('Name', TextBox_Label6.value), ('Axis', Toggle_Axis6.value), ('Color', ColorPicker_Color6.value), ('Type', Toggle_Type6.value)])
    Signal_7 = dict([('Signal', DropDown_Signal7.value), ('Name', TextBox_Label7.value), ('Axis', Toggle_Axis7.value), ('Color', ColorPicker_Color7.value), ('Type', Toggle_Type7.value)])
    Signal_8 = dict([('Signal', DropDown_Signal8.value), ('Name', TextBox_Label8.value), ('Axis', Toggle_Axis8.value), ('Color', ColorPicker_Color8.value), ('Type', Toggle_Type8.value)])
    Signal_9 = dict([('Signal', DropDown_Signal9.value), ('Name', TextBox_Label9.value), ('Axis', Toggle_Axis9.value), ('Color', ColorPicker_Color9.value), ('Type', Toggle_Type9.value)])
    return(Signal_1,Signal_2, Signal_3,Signal_4,Signal_5,Signal_6,Signal_7,Signal_8,Signal_9)

def UpdatePlotForm(TextBox_Title, TextBox_XLabel, TextBox_YLabel, IntSlider_Height, IntSlider_Width, DropDown_Theme,CheckBox_DataTable):
    
    PlotForm = dict([('title', TextBox_Title.value), ('xlabel', TextBox_XLabel.value), ('ylabel', TextBox_YLabel.value), ('height', IntSlider_Height.value), ('width', IntSlider_Width.value), ('theme', DropDown_Theme.value), ('print_signals', CheckBox_DataTable.value)])
    #export_html = CheckBox_ExportHTML.value
    #export_csv = CheckBox_ExportCSV.value
    
    return(PlotForm)

def MainPlot(Y_s, PlotForm, df):

    plot_df = pd.DataFrame(Y_s).copy()
    plot_df = plot_df[plot_df["Signal"]!='Not Selected']
    plot_df.reset_index(inplace=True)
    
    # Plot
    plot = go.Figure()
    
    Type = []
    
    for row in range(0,len(plot_df)):
    
        if plot_df["Name"][row] != str():
            Name = plot_df["Name"][row]
        else:
            Name = plot_df["Signal"][row]
        
        if plot_df["Type"][row] == 'lines':
                plot.add_trace	(go.Scatter (  
                                    x       		= Index,
                                    y       		= df[plot_df["Signal"][row]],
                                    name 			= Name,
                                    hovertemplate 	= '%{y:.2f} ',
                                    mode            = 'lines',
                                    line            =dict(color=plot_df["Color"][row])
                        )           )
                
        elif plot_df["Type"][row] == 'markers':
                plot.add_trace	(go.Scatter (  
                                    x       		= Index,
                                    y       		= df[plot_df["Signal"][row]],
                                    name 			= Name,
                                    hovertemplate 	= '%{y:.2f} ',
                                    mode            = 'markers',
                                    marker          = dict(color=plot_df["Color"][row])                              
                        )           )
                
        elif plot_df["Type"][row] == 'lines+markers':
                plot.add_trace	(go.Scatter (  
                                    x       		= Index,
                                    y       		= df[plot_df["Signal"][row]],
                                    name 			= Name,
                                    hovertemplate 	= '%{y:.2f} ',
                                    mode            = 'lines+markers',
                                    marker          = dict(color=plot_df["Color"][row]),
                                    line            = dict(color=plot_df["Color"][row])
                        )           )
    
            
    plot.update_layout	(
                        title 		= PlotForm["title"],
                        xaxis_title = PlotForm["xlabel"],
                        yaxis_title	= PlotForm["ylabel"],
                        hovermode	= "x",
                        template	= PlotForm["theme"],
                        autosize 	= False,
                        height 		= PlotForm["height"],
                        width 		= PlotForm["width"]
                        )
    
    '''
    to do add opcity, and marker styles.
    '''
    return(plot.show(), display(plot_df))

def on_button_clicked(b):
    with output:
        clear_output()
        Y_s = UpdateSignals(DropDown_Signal1, DropDown_Signal2,DropDown_Signal3,DropDown_Signal4,DropDown_Signal5,DropDown_Signal6,DropDown_Signal7,DropDown_Signal8,DropDown_Signal9,TextBox_Label1,TextBox_Label2,TextBox_Label3,TextBox_Label4,	TextBox_Label5,	TextBox_Label6,	TextBox_Label7,	TextBox_Label8,	TextBox_Label9,	Toggle_Axis1,	Toggle_Axis2,	Toggle_Axis3,	Toggle_Axis4,	Toggle_Axis5,	Toggle_Axis6,	Toggle_Axis7,	Toggle_Axis8,	Toggle_Axis9,	ColorPicker_Color1,	ColorPicker_Color2,	ColorPicker_Color3,ColorPicker_Color4,ColorPicker_Color5,ColorPicker_Color6,ColorPicker_Color7,ColorPicker_Color8,ColorPicker_Color9,Toggle_Type1,Toggle_Type2,Toggle_Type3,Toggle_Type4,Toggle_Type5,Toggle_Type6,Toggle_Type7,Toggle_Type8,Toggle_Type9)
        PlotForm = UpdatePlotForm(TextBox_Title, TextBox_XLabel, TextBox_YLabel, IntSlider_Height, IntSlider_Width, DropDown_Theme,CheckBox_DataTable)
        MainPlot(Y_s, PlotForm, df)
        
button = widgets.Button(description="Generate Plot")
output = widgets.Output()

# Create dataframe from all_files csv's
input_file = list(Button_Upload.value.values())[0]
content = input_file['content']
content = io.StringIO(content.decode('utf-8'))
df = pd.read_csv(content)

signals = list(df)
signals.insert(0, "Not Selected")
signals.insert(1, "Index")
Index = np.arange(0, len(df),1)

Y_Axis = ['Y1','Y2']
Types = ['lines','markers','lines+markers']
Type_Icons = ['slash','circle','project-diagram']

'''
Signal: Signal selector for each signal
'''
DropDown_Signal0 = widgets.Dropdown(
    options=signals,
    value=signals[1],
    description='X Axis:',
    disabled=False,
)

DropDown_Signal1 = widgets.Dropdown(
    options=signals,
    value=signals[2],
    description='Signal 1:',
    disabled=False,
)

DropDown_Signal2 = widgets.Dropdown(
    options=signals,
    value='Not Selected',
    description='Signal 2:',
    disabled=False,
)

DropDown_Signal3 = widgets.Dropdown(
    options=signals,
    value='Not Selected',
    description='Signal 3:',
    disabled=False,
)

DropDown_Signal4 = widgets.Dropdown(
    options=signals,
    value='Not Selected',
    description='Signal 4:',
    disabled=False,
)

DropDown_Signal5 = widgets.Dropdown(
    options=signals,
    value='Not Selected',
    description='Signal 5:',
    disabled=False,
)

DropDown_Signal6 = widgets.Dropdown(
    options=signals,
    value='Not Selected',
    description='Signal 6:',
    disabled=False,
)

DropDown_Signal7 = widgets.Dropdown(
    options=signals,
    value='Not Selected',
    description='Signal 7:',
    disabled=False,
)

DropDown_Signal8 = widgets.Dropdown(
    options=signals,
    value='Not Selected',
    description='Signal 8:',
    disabled=False,
)

DropDown_Signal9 = widgets.Dropdown(
    options=signals,
    value='Not Selected',
    description='Signal 9:',
    disabled=False,
)

'''
Rename Label: option to rename labels
'''
TextBox_Label0 = widgets.Text(
    value='',
    placeholder='(Optional)',
    description='Rename X:',
    disabled=False
)

TextBox_Label1 = widgets.Text(
    value='',
    placeholder='(Optional)',
    description='Rename 1:',
    disabled=False
)

TextBox_Label2 = widgets.Text(
    value='',
    placeholder='(Optional)',
    description='Rename 2:',
    disabled=False
)
TextBox_Label3 = widgets.Text(
    value='',
    placeholder='(Optional)',
    description='Rename 3:',
    disabled=False
)
TextBox_Label4 = widgets.Text(
    value='',
    placeholder='(Optional)',
    description='Rename 4:',
    disabled=False
)
TextBox_Label5 = widgets.Text(
    value='',
    placeholder='(Optional)',
    description='Rename 5:',
    disabled=False
)
TextBox_Label6 = widgets.Text(
    value='',
    placeholder='(Optional)',
    description='Rename 6:',
    disabled=False
)
TextBox_Label7 = widgets.Text(
    value='',
    placeholder='(Optional)',
    description='Rename 7:',
    disabled=False
)
TextBox_Label8 = widgets.Text(
    value='',
    placeholder='(Optional)',
    description='Rename 8:',
    disabled=False
)
TextBox_Label9 = widgets.Text(
    value='',
    placeholder='(Optional)',
    description='Rename 9:',
    disabled=False
)
'''
Axis: Axis selector for each signal
'''

Toggle_Axis1 = widgets.ToggleButtons(
    options=Y_Axis,
    description='Axis 1:',
    disabled=False,
    button_style=''
)

Toggle_Axis2 = widgets.ToggleButtons(
    value='Y1',
    options=Y_Axis,
    description='Axis 2:',
    disabled=False,
    button_style=''
)
Toggle_Axis3 = widgets.ToggleButtons(
    value='Y1',
    options=Y_Axis,
    description='Axis 3:',
    disabled=False,
    button_style=''
)
Toggle_Axis4 = widgets.ToggleButtons(
    value='Y1',
    options=Y_Axis,
    description='Axis 4:',
    disabled=False,
    button_style=''
)
Toggle_Axis5 = widgets.ToggleButtons(
    value='Y1',
    options=Y_Axis,
    description='Axis 5:',
    disabled=False,
    button_style=''
)
Toggle_Axis6 = widgets.ToggleButtons(
    value='Y1',
    options=Y_Axis,
    description='Axis 6:',
    disabled=False,
    button_style=''
)
Toggle_Axis7 = widgets.ToggleButtons(
    value='Y1',
    options=Y_Axis,
    description='Axis 7:',
    disabled=False,
    button_style=''
)

Toggle_Axis8 = widgets.ToggleButtons(
    value='Y1',
    options=Y_Axis,
    description='Axis 8:',
    disabled=False,
    button_style=''
)

Toggle_Axis9 = widgets.ToggleButtons(
    value='Y1',
    options=Y_Axis,
    description='Axis 9:',
    disabled=False,
    button_style=''
)

'''
Colors: Color picker for each signal
'''
blue 	= plotly.colors.qualitative.Plotly[0]
red 	= plotly.colors.qualitative.Plotly[1]
green 	= plotly.colors.qualitative.Plotly[2]
purple 	= plotly.colors.qualitative.Plotly[3]
orange 	= plotly.colors.qualitative.Plotly[4]
cyan 	= plotly.colors.qualitative.Plotly[5]
rose 	= plotly.colors.qualitative.Plotly[6]
lime    = plotly.colors.qualitative.Plotly[7]
pink    = plotly.colors.qualitative.Plotly[8]
sand    = plotly.colors.qualitative.Plotly[9]

ColorPicker_Color1 = widgets.ColorPicker(
    concise=True,
    description='Color 1: ',
    value=blue,
    disabled=False
)

ColorPicker_Color2 = widgets.ColorPicker(
    concise=True,
    description='Color 2: ',
    value=red,
    disabled=False
)

ColorPicker_Color3 = widgets.ColorPicker(
    concise=True,
    description='Color 3: ',
    value=green,
    disabled=False
)

ColorPicker_Color4 = widgets.ColorPicker(
    concise=True,
    description='Color 4: ',
    value=purple,
    disabled=False
)

ColorPicker_Color5 = widgets.ColorPicker(
    concise=True,
    description='Color 5: ',
    value=orange,
    disabled=False
)

ColorPicker_Color6 = widgets.ColorPicker(
    concise=True,
    description='Color 6: ',
    value=cyan,
    disabled=False
)

ColorPicker_Color7 = widgets.ColorPicker(
    concise=True,
    description='Color 7: ',
    value=rose,
    disabled=False
)

ColorPicker_Color8 = widgets.ColorPicker(
    concise=True,
    description='Color 8: ',
    value=lime,
    disabled=False
)

ColorPicker_Color9 = widgets.ColorPicker(
    concise=True,
    description='Color 9: ',
    value=pink,
    disabled=False
)

'''
Type: select plot type for each signal
'''

Toggle_Type1 = widgets.ToggleButtons(
    options=Types,
    description='Type 1:',
    disabled=False,
    button_style=''
    #icons=Type_Icons
)

Toggle_Type2 = widgets.ToggleButtons(
    options=Types,
    description='Type 2:',
    disabled=False,
    button_style=''
    #icons=Type_Icons
)

Toggle_Type3 = widgets.ToggleButtons(
    options=Types,
    description='Type 3:',
    disabled=False,
    button_style=''
    #icons=Type_Icons
)

Toggle_Type4 = widgets.ToggleButtons(
    options=Types,
    description='Type 4:',
    disabled=False,
    button_style=''
    #icons=Type_Icons
)

Toggle_Type5 = widgets.ToggleButtons(
    options=Types,
    description='Type 5:',
    disabled=False,
    button_style=''
    #icons=Type_Icons
)

Toggle_Type6 = widgets.ToggleButtons(
    options=Types,
    description='Type 6:',
    disabled=False,
    button_style=''
    #icons=Type_Icons
)

Toggle_Type7 = widgets.ToggleButtons(
    options=Types,
    description='Type 7:',
    disabled=False,
    button_style=''
    #icons=Type_Icons
)

Toggle_Type8 = widgets.ToggleButtons(
    options=Types,
    description='Type 8:',
    disabled=False,
    button_style=''
    #icons=Type_Icons
)

Toggle_Type9 = widgets.ToggleButtons(
    options=Types,
    description='Type 9:',
    disabled=False,
    button_style=''
    #icons=Type_Icons
)

'''
Style: Setup grid to display  widgets
'''

TextBox_Title = widgets.Text(
    value='',
    placeholder='Plot Title Label',
    description='Plot Title',
    disabled=False
)

TextBox_XLabel = widgets.Text(
    value='',
    placeholder='Horizontal Axis Label',
    description='X Label',
    disabled=False
)

TextBox_YLabel = widgets.Text(
    value='',
    placeholder='Left Vertical Axis Label',
    description='Y Label',
    disabled=False
)

IntSlider_Height = widgets.IntSlider(
    value=800,
    min=0,
    max=2000,
    step=1,
    description='Height',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

IntSlider_Width = widgets.IntSlider(
    value=1200,
    min=0,
    max=3000,
    step=1,
    description='Width',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

DropDown_Theme = widgets.Dropdown(
    options=[('White', 'plotly_white'), ('Dark', 'plotly_dark'), ('GGplot2', 'ggplot2'), ('Seaborn', 'seaborn'),('Presentation', 'presentation')],
    value='plotly_white',
    description='Theme',
    disabled=False,
)

CheckBox_DataTable = widgets.Checkbox(
    value=False,
    description='Display results in table',
    disabled=False,
    indent=True
)


CheckBox_ExportHTML = widgets.Checkbox(
    value=False,
    description='Export to HTML',
    disabled=True,
    indent=True
)

CheckBox_ExportCSV = widgets.Checkbox(
    value=False,
    description='Export table to CSV',
    disabled=True,
    indent=True
)


'''
Grid: Setup grid to display  widgets
'''
Grid_Height = 18
Grid_Width = 7

grid = GridspecLayout(Grid_Height, Grid_Width)

grid[0, 0] = DropDown_Signal0
grid[1, 0] = DropDown_Signal1
grid[2, 0] = DropDown_Signal2
grid[3, 0] = DropDown_Signal3
grid[4, 0] = DropDown_Signal4
grid[5, 0] = DropDown_Signal5
grid[6, 0] = DropDown_Signal6
grid[7, 0] = DropDown_Signal7
grid[8, 0] = DropDown_Signal8
grid[9, 0] = DropDown_Signal9

grid[0, 1] = TextBox_Label0
grid[1, 1] = TextBox_Label1
grid[2, 1] = TextBox_Label2
grid[3, 1] = TextBox_Label3
grid[4, 1] = TextBox_Label4
grid[5, 1] = TextBox_Label5
grid[6, 1] = TextBox_Label6
grid[7, 1] = TextBox_Label7
grid[8, 1] = TextBox_Label8
grid[9, 1] = TextBox_Label9

grid[1, 2] = Toggle_Axis1
grid[2, 2] = Toggle_Axis2
grid[3, 2] = Toggle_Axis3
grid[4, 2] = Toggle_Axis4
grid[5, 2] = Toggle_Axis5
grid[6, 2] = Toggle_Axis6
grid[7, 2] = Toggle_Axis7
grid[8, 2] = Toggle_Axis8
grid[9, 2] = Toggle_Axis9

grid[1, 3] = ColorPicker_Color1
grid[2, 3] = ColorPicker_Color2
grid[3, 3] = ColorPicker_Color3
grid[4, 3] = ColorPicker_Color4
grid[5, 3] = ColorPicker_Color5
grid[6, 3] = ColorPicker_Color6
grid[7, 3] = ColorPicker_Color7
grid[8, 3] = ColorPicker_Color8
grid[9, 3] = ColorPicker_Color9

grid[1, 4] = Toggle_Type1
grid[2, 4] = Toggle_Type2
grid[3, 4] = Toggle_Type3
grid[4, 4] = Toggle_Type4
grid[5, 4] = Toggle_Type5
grid[6, 4] = Toggle_Type6
grid[7, 4] = Toggle_Type7
grid[8, 4] = Toggle_Type8
grid[9, 4] = Toggle_Type9

grid[11,0] = TextBox_Title
grid[12,0] = TextBox_XLabel
grid[13,0] = TextBox_YLabel
grid[11,1] = IntSlider_Width
grid[12,1] = IntSlider_Height
grid[13,1] = DropDown_Theme
grid[14,0] = CheckBox_DataTable
grid[15,0] = CheckBox_ExportHTML
grid[16,0] = CheckBox_ExportCSV

grid
```
